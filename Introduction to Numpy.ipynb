## Accessing the elements of multi-Dimensional Array

```python
lstOfLists[0,2]
```

```python
lstOfLists[0,2] = 100
lstOfLists
```

```python
columnFirst = [0,0,0]
columnFirst
```

```python
lstOfLists
```

```python
lstOfLists[:,1] = columnFirst
lstOfLists
```

# Randomly generated arrays

```python
# If you need to generate Arrays with Random numbers
np.random.rand(5,2)
```

```python
# The above cell generates array with Random numbers, 
# Everytime the above cell is executed the elements of the array are different.
# Every user running the above cell might get unique elements
# TO keep the elements constant across multiple users and across multiple runs. 
# set seed funcction is used. 

np.random.seed(2)
np.random.rand(5,2)
```

```python

np.random.seed(2)
np.random.randint(low=0, high =100, size=(5,2))
```
